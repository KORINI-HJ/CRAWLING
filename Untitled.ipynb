```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome("./chromedriver")
driver.get("http://www.seoul.go.kr/coronaV/coronaStatus.do")
time.sleep(2)
dsc_hj = []
```

```python
driver.find_element_by_xpath('//*[@id="container"]/div[2]/div/div[6]/ul/li[2]/button').click()
time.sleep(3)
```

```python
for i in range(0, 150, 2):
    dsc_hj.append(driver.find_elements_by_xpath('//*[@id="DataTables_Table_0"]/tbody/tr['+str(i)+']/td/p'))
```

```python
for i in dsc_hj:
    for j in i:
        print(j.text)
```

```python
data = map(str, input(j.text).split())
```
